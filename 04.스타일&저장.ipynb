```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #한글 폰트
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 이용시 축 음수표시
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y, linewidth = 5)
```

```python
plt.plot(x, y, marker='o')
```

```python
plt.plot(x, y, marker ='v', markersize = 10, linestyle = 'None')
```

```python
plt.plot(x, y, marker = 'X', markersize = 20, markeredgecolor='red', markerfacecolor='green')  
#matplotlib 사이트에서 값 확인 가능
```

```python
plt.plot(x, y, linestyle = ':')
```

```python
plt.plot(x, y, linestyle = '-.')
```

```python
plt.plot(x, y, color = 'green')
```

```python
plt.plot(x, y, 'rv--') #color, marker, linestyle
```

```python
plt.plot(x, y, 'go')
#markeredgecolor = mec로 사용가능, 여러 축약어 존재 사이트에서 확인
```

```python
plt.plot(x, y, marker = 'o', mfc = 'r', ms = 10, alpha = 0.2) #alpha는 투명도 (0~1)
```

```python
plt.figure(figsize=(10,5), dpi = 200) #dots per inch, 크기 조절
plt.plot(x, y)
```

```python
plt.figure(facecolor = '#5691f0')
plt.plot(x, y)
plt.savefig('graph.png', dpi = 100) #파일로 저장
```
