```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #한글 폰트
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 이용시 축 음수표시
```

```python
x = [1, 2, 3]
y = [2, 4, 6]
```

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend(loc = 'lower right')  #matplot legend 들어가서 설정 확인 (0.5, 0.4)와 같은 표현으로 설정도 가능
```
