```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #한글 폰트
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 이용시 축 음수표시
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.title('꺾은선 그래프', fontdict={'family': 'HYGungSo-Bold'})
```

```python
plt.xlabel('X축', color='blue', loc = 'right') #left, center, right
plt.ylabel('Y축', loc = 'top') #top, center, bottom
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
plt.show()
```
